```python
# TASK 4: BASIC RULE-BASED CHATBOT
# ----------------------------

print("Welcome! I am a simple chatbot. Type 'bye' if want to exit.\n")

# Step 1: Create a dictionary of predefined responses
responses = {
    "hello": "Hi there!",
    "hi": "Hello!",
    "how are you": "I'm doing great! Thanks for asking.",
    "bye": "Goodbye! Have a nice day!"
}

# Step 2: Start an infinite loop so chatbot keeps talking
while True:
    # Step 3: Take input from the user
    user_input = input("You: ").strip().lower()

    # Step 4: Check if user typed something empty
    if user_input == "":
        print("Bot: Please type something...")
        continue

    # Step 5: Check if the message matches our known responses
    if user_input in responses:
        # Step 6: Print the matched response
        print("Bot:", responses[user_input])

        # Step 7: If user says 'bye', exit the loop
        if user_input == "bye":
            break
    else:
        # Step 8: Fallback reply if message not understood
        print("Bot: Sorry, I didn't understand that. Try saying 'hello' or 'how are you'.")
```
