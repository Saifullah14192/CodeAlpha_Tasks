```python
# 1) import random module to pick random word from list
import random  # used to choose a random secret word

# 2) 5 predefined words
words = ["python", "laptop", "banana", "school", "coding"]  # list of words player can get

# 3) choose one random secret word
secret_word = random.choice(words)  # picks one string from 'words'

# 4) prepare display: underscores for each letter of the secret word
display = ["_"] * len(secret_word)  # creates list like ['_', '_', '_']

# 5) keep track of letters already tried
guessed_letters = set()  # set for fast 'in' checks and to avoid duplicates

# 6) wrong attempts counter and max allowed wrong guesses
wrong_attempts = 0
max_attempts = 6

# 7) welcome message and current display shown to player
print("< Welcome to Hangman >!!")
print("Guess the secret word:", " ".join(display))  # join display with spaces

# 8) main game loop: continue while wrong attempts less than max and there are underscores left
while wrong_attempts < max_attempts and "_" in display:
    # 9) ask player for a single letter (normalize to lowercase)
    guess = input("Enter a single letter: ").strip().lower()

    # 10) validate input: must be exactly 1 alphabetic character
    if len(guess) != 1 or not guess.isalpha():
        print("Please enter a single letter (a-z).")
        continue  # go back to start of while loop

    # 11) if player already tried this letter, inform and skip
    if guess in guessed_letters:
        print("You already tried this letter.")
        continue

    # 12) add letter to guessed_letters set
    guessed_letters.add(guess)

    # 13) if the guessed letter is in the secret word, reveal it in 'display'
    if guess in secret_word:
        for i, ch in enumerate(secret_word):  # check each position
            if ch == guess:
                display[i] = guess  # reveal the letter
        print("Good!!!!", " ".join(display))
    else:
        # 14) wrong guess: increment wrong attempts and show progress
        wrong_attempts += 1
        print(f"Wrong!!! {wrong_attempts}/{max_attempts} attempts used.")
        print(" ".join(display))

# 15) after loop, determine win or loss
if "_" not in display:
    print("\n Congratulations>> — you guessed the word:", secret_word)
else:
    print("\n Game over>!!. The word was:", secret_word)
```
